```python
# Softmax function
def softmax(EUs):
    """Compute the softmax of vector EUs."""
    e_EUs = np.exp(EUs - np.max(EUs))  # Subtract max for numerical stability
    return e_EUs / e_EUs.sum(axis=0)
```
