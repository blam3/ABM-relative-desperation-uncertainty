# Modified Decision-makiung model under uncertainty with softmax desperation threshold

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def u_shaped_softmax(logits, b):
    modified_logits = b * (logits ** 2)
    return np.exp(modified_logits) / np.sum(np.exp(modified_logits))
```

```python

# Input logits
logits_u_shape = np.linspace(-3, 3, 100)

# Parameter (scaling factor)
b = 1

# Compute U-shaped softmax
u_shaped_softmax_values = u_shaped_softmax(logits_u_shape, b)
u_shaped_softmax_values
```

```python
plt.plot(logits_u_shape, u_shaped_softmax_values, marker='o')
plt.title('Modified Softmax Function')
plt.xlabel('Resources')
plt.ylabel('Softmax Values')
plt.grid()
```
